# Evaluacion de KPI en la industria Aérea:

Los KPI son una herramienta importante para la gestión de la seguridad aérea. Al medir el rendimiento de los sistemas de seguridad, los KPI ayudan a las organizaciones a identificar áreas de mejora y tomar medidas para reducir el riesgo de accidentes.


## 1. KPI: Análisis de fatalidad de la Tripulación

## Objetivo:
- Evaluar la disminución de un 10% en la tasa de fatalidad de la tripulación en los últimos 10 años, comparado con la década anterior.

## Nombre del KPI:
- Tasa de Fatalidad de la Tripulación (Comparación Decadal)

## Forma de Medida:
La tasa de fatalidad de la tripulación se mide calculando el número total de tripulantes fallecidos en accidentes registrados durante un período de 10 años y dividiéndolo por la cantidad total de accidentes aéreos ocurridos en ese mismo período de tiempo. La fórmula para calcular esta tasa es:

Tasa de Fatalidad de la Tripulación = (Suma total de fallecidos en el período de 10 años) / (Suma total de accidentes en el período de 10 años)

## Meta:
La meta es lograr una disminución del 10% en la tasa de fatalidad de la tripulación en los últimos 10 años en comparación con la década anterior. Esto significa que la tasa de fatalidad de la tripulación en los últimos 10 años debe ser un 10% menor que la tasa de fatalidad de la tripulación de la década anterior. La meta se expresa como una reducción del 10% en la tasa en términos porcentuales.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame llamado 'accidentes' ya cargado.

# Especifica la ruta del archivo CSV que deseas leer
archivo_csv = 'accidentes.csv'

# Usa pd.read_csv() para leer el archivo CSV y convertirlo en un DataFrame
accidentes= pd.read_csv(archivo_csv)
```

```python
# Agrupa los accidentes por año y calcula la suma de tripulación a bordo y tripulación fallecida por año
tasa_fatalidad_tripulacion = accidentes.groupby('anio')[['tripulacion_a_bordo', 'tripulacion_fallecida']].sum()

# Calcula la tasa de fatalidad de la tripulación por año
tasa_fatalidad_tripulacion['tasa_fatalidad'] = (tasa_fatalidad_tripulacion['tripulacion_fallecida'] / tasa_fatalidad_tripulacion['tripulacion_a_bordo']) * 100
```

```python
# Filtra los datos para los últimos 10 años
current_decade = accidentes[accidentes['anio'] >= (max(accidentes['anio']) - 10)]

# Filtra los datos para la década anterior
previous_decade = accidentes[(accidentes['anio'] >= (max(accidentes['anio']) - 20)) & (accidentes['anio'] < (max(accidentes['anio']) - 10))]

# Calcula la tasa de fatalidad de la tripulación para ambas décadas
current_decade_fatalities = current_decade['tripulacion_fallecida'].sum()
current_decade_tripulacion_total = current_decade['tripulacion_a_bordo'].sum()
current_decade_fatality_rate = (current_decade_fatalities / current_decade_tripulacion_total) * 100

previous_decade_fatalities = previous_decade['tripulacion_fallecida'].sum()
previous_decade_tripulacion_total = previous_decade['tripulacion_a_bordo'].sum()
previous_decade_fatality_rate = (previous_decade_fatalities / previous_decade_tripulacion_total) * 100

# Calcula la reducción porcentual en la tasa de fatalidad de la tripulación
reduction_percentage = ((previous_decade_fatality_rate - current_decade_fatality_rate) / previous_decade_fatality_rate) * 100

# Imprime los resultados
print("Tasa de Fatalidad de la Tripulación en la Década Actual:", current_decade_fatality_rate, "%")
print("Tasa de Fatalidad de la Tripulación en la Década Anterior:", previous_decade_fatality_rate, "%")
print("Reducción Porcentual en la Tasa de Fatalidad de la Tripulación:", reduction_percentage, "%")

# Verifica si se ha logrado la reducción del 10%
if reduction_percentage >= 10:
    print("Se ha logrado una reducción del 10% o más en la tasa de fatalidad de la tripulación.")
else:
    print("No se ha logrado una reducción del 10% en la tasa de fatalidad de la tripulación.")
```

```python
# Crear un gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(tasa_fatalidad_tripulacion.index, tasa_fatalidad_tripulacion['tasa_fatalidad'], marker='o', linestyle='-')
plt.title('Tasa de Fatalidad de la Tripulación por Año')
plt.xlabel('Año')
plt.ylabel('Tasa de Fatalidad (%)')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Tasas de fatalidad y décadas
decades = ['Década Anterior', 'Década Actual']
fatalities_rates = [previous_decade_fatality_rate, current_decade_fatality_rate]

# Crear el gráfico de barras
plt.bar(decades, fatalities_rates, color=['orange', 'green'])
plt.xlabel('Décadas')
plt.ylabel('Tasa de Fatalidad de la Tripulación (%)')
plt.title('Tasa de Fatalidad de la Tripulación en las Décadas Anterior y Actual')

# Agregar la reducción porcentual como anotación en el gráfico
plt.text(0.5, 10, f'Reducción Porcentual: {reduction_percentage:.2f}%', ha='center', va='bottom', fontsize=12, color='red')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```


El análisis de las tasas de fatalidad de la tripulación en las dos últimas décadas indica que la seguridad de la tripulación en accidentes aéreos no ha mejorado en la década actual en comparación con la década anterior.

En la década actual, la tasa de fatalidad de la tripulación fue del 88.63%, lo que indica que un alto porcentaje de accidentes aéreos en esta década resultaron en fatalidades entre la tripulación. En la década anterior, la tasa de fatalidad de la tripulación fue del 82.89%, lo que muestra que hubo una tasa ligeramente menor de fatalidades entre la tripulación en la década anterior.

La reducción porcentual en la tasa de fatalidad de la tripulación entre la década anterior y la década actual es del -6.92%, lo que indica que, en lugar de haber una mejora en la seguridad de la tripulación, ha habido un aumento en la tasa de fatalidad en la década actual en comparación con la década anterior.

Este aumento es significativo, ya que sugiere que los esfuerzos de seguridad no han tenido éxito en reducir las fatalidades en la tripulación de manera significativa. Esto subraya la necesidad de investigar y abordar las causas de los accidentes aéreos y tomar medidas efectivas para mejorar la seguridad de la tripulación en la industria de la aviación.

Algunas posibles causas del aumento en la tasa de fatalidad de la tripulación en la década actual incluyen:

- El aumento de la complejidad de los aviones, que dificulta la recuperación de accidentes.
- El cambio climático, que puede provocar condiciones meteorológicas adversas que pueden provocar accidentes.

## 2. KPI: Eficiencia de la seguridad Aérea

## KPI: Tasa de Supervivencia de Pasajeros

## Objetivo:
- Evaluar la disminución de un 5% en la tasa de supervivencia de pasajeros en accidentes aéreos en los ultimos 5 años, comparado con los 5 años anterior.


## Forma de Medida:

- Tasa de Supervivencia de Pasajeros = (Número de pasajeros sobrevivientes) / (Número total de pasajeros a bordo)

## Meta:
- La meta es lograr disminuir el 5% en la tasa de supervivencia en los ultimos 5 años en comparación con los 5 años anterior.


```python
# Agrupa los datos por año y calcula la tasa de sobrevivencia en porcentaje
tasa_sobrevivencia_por_anio = (accidentes.groupby('anio')['sobrevivientes'].sum() / accidentes.groupby('anio')['total_ocupantes'].sum()) * 100

# Agrupa los datos por año y calcula la tasa de mortalidad en porcentaje
tasa_mortalidad_por_anio = 100 - ((accidentes.groupby('anio')['sobrevivientes'].sum() / accidentes.groupby('anio')['total_ocupantes'].sum()) * 100)
```

```python

# Crea un gráfico de líneas para la tasa de supervivencia
plt.figure(figsize=(10, 6))
plt.plot(tasa_sobrevivencia_por_anio.index, tasa_sobrevivencia_por_anio.values, label='Tasa de Supervivencia', marker='o')

# Crea un gráfico de líneas para la tasa de mortalidad
plt.plot(tasa_mortalidad_por_anio.index, tasa_mortalidad_por_anio.values, label='Tasa de Mortalidad', marker='x')

# Configura el título y etiquetas del eje x y y
plt.title('Tasa de Supervivencia y Mortalidad por Año')
plt.xlabel('Año')
plt.ylabel('Porcentaje')

# Agrega una leyenda
plt.legend()

# Muestra el gráfico
plt.grid(True)
plt.show()
```

```python
# Filtrar los datos para los últimos 5 años
ultimo_anio = accidentes['anio'].max()
primer_anio = ultimo_anio - 4

# Calcular la tasa de supervivencia para los últimos 5 años
tasa_supervivencia_ultimos_5_anios = (accidentes[(accidentes['anio'] >= primer_anio) & (accidentes['anio'] <= ultimo_anio)]['sobrevivientes'].sum() / 
                                      accidentes[(accidentes['anio'] >= primer_anio) & (accidentes['anio'] <= ultimo_anio)]['total_ocupantes'].sum()) * 100

# Calcular la tasa de supervivencia para los 5 años anteriores
tasa_supervivencia_anteriores = (accidentes[(accidentes['anio'] >= primer_anio - 5) & (accidentes['anio'] <= primer_anio - 1)]['sobrevivientes'].sum() / 
                                 accidentes[(accidentes['anio'] >= primer_anio - 5) & (accidentes['anio'] <= primer_anio - 1)]['total_ocupantes'].sum()) * 100

# Calcular la diferencia en la tasa de supervivencia
diferencia_tasa = tasa_supervivencia_anteriores - tasa_supervivencia_ultimos_5_anios

# Imprimir los resultados en porcentaje
print("Tasa de supervivencia de los últimos 5 años:", tasa_supervivencia_ultimos_5_anios, "%")
print("Tasa de supervivencia de los 5 años anteriores:", tasa_supervivencia_anteriores, "%")
print("Diferencia en la tasa de supervivencia:", diferencia_tasa, "%")

# Verificar si la disminución es del 5% o más
if diferencia_tasa >= 5.0:
    print("Se ha logrado una disminución del 5% o más en la tasa de supervivencia.")
else:
    print("No se ha logrado una disminución del 5% en la tasa de supervivencia.")
```

 El objetivo de la industria de la aviación es reducir la tasa de supervivencia en un 5%. Sin embargo, los resultados muestran que la tasa de supervivencia ha disminuido en un 2.93 %, lo que está por debajo del objetivo.

Ante tal desviación se pueden considerar medidas como:

- Mejorar la seguridad de los aviones. Esto podría incluir el desarrollo de nuevos materiales y estructuras que sean más resistentes a los daños.
- Mejorar la formación de los pilotos y los controladores aéreos. Esto podría incluir el entrenamiento en situaciones de emergencia.
- Mejorar la concienciación sobre la seguridad entre los pasajeros. Esto podría incluir la capacitación en seguridad para los pasajeros.
Tomando estas medidas, se puede reducir la tasa de fatalidad en accidentes aéreos y mejorar la seguridad en la industria de la aviación.
